```python
import tensorflow as tf
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
import Prediction as p
```

```python
#Load Data
m = np.load('Data_AD.npy')
```

```python
#Import Physics-aware Model-Order Reduction Model
import sys  
sys.path.insert(0, '../Training/')

import model as mor
```

```python
latent_dim = 5
model = mor.MODEL(latent_dim)
```

```python
#Load weights
model.load_weights('weights_AD')
```

```python
#Reconstruction of densities.
x=np.linspace(-1,1,25)
sample=15 #Sample
time=20 # Time step
density_m,density_s=p.prediction(sample,time,model)
plt.plot(x,density_m,label='Posterior Mean')
plt.fill_between(x,(density_m-1*density_s),(density_m+1*density_s),alpha=0.3,label='+/- 1 Standard Deviation')
plt.plot(x,m[sample,time,:]/250000,label='Reference')
plt.axis([-1,1,0.02,0.06])
plt.legend()
plt.xlabel('$s$')
plt.ylabel(r'$\rho(s,t)$')
```

```python
lamb_r=model.Theta_r.numpy()  #Real part of lambda= -lamb_r**2
lamb_i=model.Theta_i.numpy()  #Imaginary part of lambda
```

```python
#Prediction for time-steps not contained in the training data
x=np.linspace(-1,1,25)
sample=15  #sample
steps=40   #number of steps
density_m,density_s=p.prediction_extrapolative(sample,steps,model,lamb_r,lamb_i)
plt.plot(x,density_m,label='Posterior Mean')
plt.fill_between(x,(density_m-1*density_s),(density_m+1*density_s),alpha=0.3,label='+/- 1 Standard Deviation')
plt.plot(x,m[sample,39+steps,:]/250000,label='Reference')
plt.axis([-1,1,0.03,0.05])
plt.legend()
plt.xlabel('$s$')
plt.ylabel(r'$\rho(s,t)$')
```

```python
#Prediction for 1000 time steps = steady state
x=np.linspace(-1,1,25)
sample=15
steps=1000
density_m,density_s=p.prediction_extrapolative(sample,steps,model,lamb_r,lamb_i)
plt.plot(x,density_m,label='Posterior Mean')
plt.fill_between(x,(density_m-1*density_s),(density_m+1*density_s),alpha=0.3,label='+/- 1 Standard Deviation')
equ=np.ones((25,))*0.04
plt.plot(x,equ,label='Reference')
plt.axis([-1,1,0.03,0.05])
plt.legend()
plt.xlabel('$s$')
plt.ylabel(r'$\rho(s,t)$')
```

```python
#Obtained values for lambda
plt.plot(-lamb_r**2,lamb_i,'rx', markersize=14)
plt.grid()
plt.xlabel('$\Re(\lambda)$')
plt.ylabel('$\Im(\lambda)$')
```

```python
#Obtained and predicted z
z_r,z_i,z_rp,z_ip=p.prediction_z(15,model,lamb_r,lamb_i)
```

```python
#Slow process
i=4
plt.plot(z_r[:,i],z_i[:,i],label='Inferred Mean')
plt.plot(z_rp[i,:],z_ip[i,:],label='Predicted Mean')
plt.legend()
plt.xlabel('$\Re(z)$')
plt.ylabel('$\Im(z)$')
```
